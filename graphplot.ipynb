```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('relationship1.csv')
data


```

```python
x = data['Move']
y = data['Width']

plt.plot(x, y, marker='+', linestyle='-',color = 'r')  
plt.xlabel('Distance')  
plt.ylabel('Width')  
plt.title('Width Graph')  

plt.show()
```

```python
x = data['Move']
y = data['Height']

plt.plot(x, y, marker='+', linestyle='-',color = 'g')  
plt.xlabel('Distance')  
plt.ylabel('Height')  
plt.title('Height Graph')  

plt.show()
```

```python
x = data['Move']
y = data['Diagonal']

plt.plot(x, y, marker='+', linestyle='-',color = 'b')  
plt.xlabel('Distance')  
plt.ylabel('Diagonal')  
plt.title('Diagonal Graph')  

plt.show()
```

```python
x = data['Move']
y = data['Diagonal']
y_2 = data['Height']
y_3 = data['Width']

plt.plot(x, y, marker='+', linestyle='-',color = 'b',label='Diagonal')  
plt.plot(x, y_2, marker='+', linestyle='-',color = 'r',label='Height')  
plt.plot(x, y_3, marker='+', linestyle='-',color = 'g',label='Width')  
plt.xlabel('')  
plt.title('Diagonal Graph')  
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# สมมติว่า 'Diagonal', 'Height', 'Width' อยู่ในรายการข้อมูล 'data'
x = data['Move']
y_diagonal = data['Diagonal']
y_height = data['Height']
y_width = data['Width']

plt.plot(x, y_diagonal, marker='+', linestyle='-', color='b', label='Diagonal')
plt.plot(x, y_height, marker='o', linestyle='--', color='g', label='Height')
plt.plot(x, y_width, marker='x', linestyle=':', color='r', label='Width')

plt.xlabel('Distance')
plt.ylabel('Value')  # คุณอาจต้องระบุชื่อแกน Y ใหม่ที่เหมาะสมกับข้อมูล
plt.title('Three Lines Graph')

plt.legend()

plt
```
