```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('relationship1.csv')
data


```

```python
x = data['Move']
y = data['Width']

plt.plot(x, y, marker='+', linestyle='-',color = 'r')  
plt.xlabel('Distance')  
plt.ylabel('Width')  
plt.title('Width Graph')  

plt.show()
```

```python
x = data['Move']
y = data['Height']

plt.plot(x, y, marker='+', linestyle='-',color = 'g')  
plt.xlabel('Distance')  
plt.ylabel('Height')  
plt.title('Height Graph')  

plt.show()
```
