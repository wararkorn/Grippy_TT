```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('robot_to_chick2.csv')
data
```

```python
x = data['distance']
y = data['width']

plt.plot(x, y, marker='+', linestyle='-',color = 'r')  
plt.xlabel('Distance')  
plt.ylabel('Width')  
plt.title('Width Graph')  

plt.show()
```

```python
x = data['distance']
y = data['height']

plt.plot(x, y, marker='+', linestyle='-',color = 'g')  
plt.xlabel('Distance')  
plt.ylabel('Height')  
plt.title('Height Graph')  

plt.show()
```

```python
x = data['distance']
y = data['diagonal']

plt.plot(x, y, marker='+', linestyle='-',color = 'g')  
plt.xlabel('Distance')  
plt.ylabel('diagonal')  
plt.title('diagonal Graph')  

plt.show()
```
